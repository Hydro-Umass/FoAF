```python
import requests

def get_noaa_stations_in_state(state_code):
    base_url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/stations"
    token = "DvlSHQkoUvbkbrPkFKZYheKZCFlUFYQe"  # Replace with your NOAA API token

    params = {
        "datasetid": "GHCND",
        "locationid": f"ST:{state_code}",
        "limit": 1000,
        "datatypeid": "TAVG,PRCP,TMAX,TMIN,AWND",
    }

    headers = {"token": token}
    
    response = requests.get(base_url, params=params, headers=headers)

    if response.status_code == 200:
        data = response.json()
        stations = data.get("results", [])
        return [station["id"] for station in stations]
    else:
        print(f"Error: {response.status_code}")
        return []

# State code for Massachusetts
ma_state_code = "MA"

# Get NOAA station codes in Massachusetts
station_codes = get_noaa_stations_in_state(ma_state_code)

# Display the list of station codes
print("NOAA Station Codes in Massachusetts:")
print(station_codes)
```

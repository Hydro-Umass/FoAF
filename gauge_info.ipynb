```python
import pandas as pd

section1info = site_no_list[:500]
chunk_size = 50
data_list2 = []

for i in range(0, len(section1info), chunk_size):
    chunk = section1info[i:i + chunk_size]
    
    for site_no in chunk:
        data = nwis.get_info(sites=site_no)
        data_list2.append(data)

sec1info = pd.concat(data_list2)
sec1info.to_csv('section1info.csv', index=False)
```

```python
import pandas as pd

section1info = site_no_list[:500]
chunk_size = 50
data_list2 = []

for i in range(0, len(section1info), chunk_size):
    chunk = section1info[i:i + chunk_size]
    
    for site_no in chunk:
        try:
            data = nwis.get_info(sites=site_no)
            # Convert the tuple to a DataFrame
            data_df = pd.DataFrame([data])
            data_list2.append(data_df)
        except Exception as e:
            print(f"Error for site {site_no}: {e}")

if data_list2:
    sec1info = pd.concat(data_list2, ignore_index=True)
    sec1info.to_csv('section1info.csv', index=False)
else:
    print("No data retrieved.")
```

```python
section1 = site_no_list[:500]

start_date = '1977-01-01'
end_date = '2006-12-31'
paramcode = '00060'

chunk_size = 50 

data_list2 = []

for i in range(0, len(section1), chunk_size):
    chunk = section1[i:i + chunk_size]
    
    for site_no in chunk:
        data = nwis.get_record(site=site_no, parameterCd=paramcode, service='dv', start=start_date, end=end_date)
        data_list2.append(data)
```

```python
sec1 = pd.concat(data_list2)
section1 = sec1.to_csv('section1.csv', index=False)
```
